```python
import re
cleaner = re.compile('(?!</?br>)<.*?>')
def cleanhtml(item):
    key, raw_html = item
    cleantext = re.sub(cleaner, '', raw_html)
    return key, cleantext
# 'https://bookthumb-phinf.pstatic.net/cover/134/575/13457579.jpg?type=m1&udate=20191029'
# 'https://bookthumb-phinf.pstatic.net/cover/134/575/13457579.jpg'
def parse_image_link(raw_link: str)->str:
    return raw_link[:raw_link.find('?')]

def split_isbn(item: dict) -> None:
    isbn10 = isbn13 = -1;
    if 'isbn' in item:
        isbn10, isbn13 = item['isbn'].split()
        del item['isbn']
    item['isbn10'] = isbn10
    item['isbn13'] = isbn13
import time
def book_updater(item: dict)-> dict:
    detail_res = item['link']
    res = requests.get(detail_res)
    time.sleep(3)
    soup = BeautifulSoup(res.text, 'html.parser')
    authorIntroContent = soup.find(id="authorIntroContent")
    
    updater = {
        "authorIntroContent": soup.find(id="authorIntroContent") if soup.find(id="authorIntroContent") else "",
        "tableOfContentsContent": soup.find(id="tableOfContentsContent").br if soup.find(id="tableOfContentsContent") else "",
        "bookIntroContent": soup.find(id="bookIntroContent").p if soup.find(id="bookIntroContent") else "",
    }
    item = dict(map(cleanhtml, item.items()))
    item['image'] = parse_image_link(item['image'])
    split_isbn(item)
    item.update(updater)
    return item
```

```python
data = map(book_updater, json.loads(res.content)['items'][:10])
```
